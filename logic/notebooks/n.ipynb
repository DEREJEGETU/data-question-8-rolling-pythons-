```python
import pickle
import pandas as pd
import re

with open("test_paras.pkl", "rb") as pk:
    pkl = pickle.load(pk)
```

```python
df = pd.DataFrame(list(pkl.items()), columns=['html_file', 'section'])
```

```python
df['access_no'] = df['html_file'].replace(r"(.+)(\.html)", r"\1", regex=True)
```

```python
df.shape
```

```python
date = re.compile(r"([A-Z][a-z]+ \d{1,2}, (\d{4}))|" +
                   "(\d{1,2} [A-Z][a-z]+, (\d{4}))|" +
                   "( (\d{4}))")

values = re.compile(r"( (\d[0-9.]+ [mb]illions?))|" +
                   "( (\d{1,3}(,\d{1,3})*) [A-Za-z]+s)|" +
                   "( (\$\d[0-9.]+ [mb]illions?))|" +
                   "( ($\d{1,3}(,\d{1,3})*) )")

keywords = ["auth", "repurchas", "total", "remain"]
```

```python
cols = df.columns.tolist()
cols.extend(['year', 'quantity'])

final = pd.DataFrame([])
```

```python
from copy import deepcopy

empty = []
for index, row in df.iterrows():
    para = row.section
    flag = 0
    sents = para.split(". ")
    for sent in sents:
        if "repurchas" not in sent or '2017' not in sent:
            continue
        if values.search(sent):
            flag = 1
            v = []
            d = []
            
            vs = values.findall(sent)
            ds = date.findall(sent)
            
            for g in vs:
                g = list(g)
                while '' in g:
                    g.remove('')
                v.append(g[1])
                
            for j in ds:
                j = list(j)
                while '' in j:
                    j.remove('')
                d.append(j[1])
                
            line = deepcopy(list(row))
            line.append(sent)
            line.append(v)
            line.append(d)
            
            final = final.append([line])
            
    if not flag:
        empty.append(row.html_file)
```
